```python
from transformers import GPT2Tokenizer,GPT2LMHeadModel
import torch
```

```python
model=GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
prompt = "the future of ai is"
input_ids = tokenizer.encode(prompt,return_tensors="pt")
```

```python
# greedy
print("greedy:")
print(tokenizer.decode(model.generate(input_ids,max_length=50,do_sample =False)[0]))
```

```python
print("Beam Search:")

output = model.generate(
    input_ids,
    max_length=50,
    num_beams=5,
    early_stopping=True
)

print(tokenizer.decode(output[0]))
```

```python
print("top k- sampling:")
output = model.generate(
    input_ids,
    max_length=50,
    do_sample=True,
    top_k=100
)

print(tokenizer.decode(output[0]))
```

```python
print("Top-P:")
output = model.generate(
    input_ids,
    max_length=50,
    do_sample=True,
    top_p=0.6
)

print(tokenizer.decode(output[0]))
```
