```python
import os
from dotenv import load_dotenv
from huggingface_hub import InferenceClient

# १. .env फाईल लोड करा
load_dotenv()
hf_token = os.getenv("hf_token")

# २. client नवीन तयार करा 
client = InferenceClient(api_key=hf_token)

# ३. Qwen मॉडेल वापरा (याला license ची गरज नाही, डायरेक्ट चालेल)
model_id = "Qwen/Qwen2.5-72B-Instruct" 

messages = [
    {"role": "user", "content": "The future of education is"}
]

# ४. API रन करा
result = client.chat_completion(
    model=model_id,
    messages=messages,
    max_tokens=50
)

# ५. उत्तर प्रिंट करा
print(result.choices[0].message.content)
```

```python
import os
from dotenv import load_dotenv
from huggingface_hub import InferenceClient

load_dotenv()
hf_token = os.getenv("hf_token")
client = InferenceClient(api_key=hf_token)

model_id = "Qwen/Qwen2.5-72B-Instruct" 
messages = [{"role": "user", "content": "The future of education is"}]

result = client.chat_completion(
    model=model_id,
    messages=messages,
    max_tokens=100
)
```
